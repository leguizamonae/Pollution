```python
pollution['Cumulative_AQI'] = pollution['O3 AQI'] + pollution['CO AQI'] + pollution['SO2 AQI'] + pollution['NO2 AQI'] 
#Se crea nueva columna con una nueva medida
```

**Se crear columna con un nuevo 칤ndice de medida acumulativo (suma del Indice de calidad de Aire de cada contaminante.)**

```python
pollution['Cumulative_AQI']=pollution['Cumulative_AQI'].astype('int16')
pollution['Cumulative_AQI']
```

**El 칤ndice de calidad del aire (AQI: Air Quality Index) posee un criterio que va de 0 a 500. Cuanto mayor sea el valor del AQI, mayor ser치 el nivel de contaminaci칩n del aire y mayor ser치 la preocupaci칩n por la salud. Por ejemplo, un valor AQI de 50 o inferior representa una buena calidad del aire, mientras que un valor AQI superior a 300 representa una calidad del aire peligrosa.
Para cada contaminante, un valor AQI de 100 generalmente corresponde a una concentraci칩n en el aire ambiental que iguala el nivel del est치ndar nacional de calidad del aire ambiental a corto plazo para la protecci칩n de la salud p칰blica. Los valores de AQI de 100 o menos generalmente se consideran satisfactorios. Cuando los valores AQI est치n por encima de 100, la calidad del aire no es saludable: al principio para ciertos grupos sensibles de personas, luego para todos a medida que los valores AQI aumentan.
El AQI se divide en seis categor칤as. Cada categor칤a corresponde a un nivel diferente de preocupaci칩n por la salud. Cada categor칤a tambi칠n tiene un color espec칤fico. El color facilita que las personas determinen r치pidamente si la calidad del aire est치 alcanzando niveles nocivos para la salud en sus comunidades:**

* **Buena: Color verde (AQI de 0 a 50).**
* **Moderada: Color amarillo (AQI de 51 a 100)**
* **Da침ina a la salud para grupos sensibles: Color naranja (AQI de 101 a 150)**
* **Da침ina a la salud: Color rojo (AQI 151 a 200)**
* **Muy da침ina a la salud: Color morado (AQI 201 a 300)**
* **Peligrosa: Color marr칩n (AQI superior a 300)**

Fuente: https://www.airnow.gov/aqi/aqi-basics-in-spanish/ 

image.png

```python
# Se crea nueva columna para categorizar valores de AQI
conditionlist = [
    (pollution['Cumulative_AQI'] > 0) & (pollution['Cumulative_AQI'] <= 50), 
    (pollution['Cumulative_AQI'] >= 51) & (pollution['Cumulative_AQI'] <= 100), 
    (pollution['Cumulative_AQI'] >= 101) & (pollution['Cumulative_AQI'] <= 150), 
    (pollution['Cumulative_AQI'] >= 151) & (pollution['Cumulative_AQI'] <= 200), 
    (pollution['Cumulative_AQI'] >= 201) & (pollution['Cumulative_AQI'] <= 300), 
    (pollution['Cumulative_AQI'] >= 301)]

choicelist = ['Buena', 'Moderada', 'Da침ina a la salud para grupos sensibles', 'Da침ina a la salud', 'Muy da침ina a la salud', 'Peligrosa']

pollution['Nivel_AQI'] = np.select(conditionlist, choicelist)
```

```python
# Se crea columna Target para clasificar niveles 
Target = {'Buena':0, 'Moderada':1, 'Da침ina a la salud para grupos sensibles':2,  'Da침ina a la salud':3, 'Muy da침ina a la salud':4, 'Peligrosa':5}
pollution['Target'] = pollution['Nivel_AQI'].map(Target)
```

Para encontrar el t칠rmino medio entre la complejidad del modelo y su ajuste a los datos, nos ayudamos de herramientas gr치ficas. Primero creando diferentes modelos, con distintos grados de complejidad y luego graficar la precisi칩n en funci칩n de la complejidad.

```python
train_prec =  []
eval_prec = []
max_deep_list = list(range(1, 15))

for deep in max_deep_list:
    arbol = DecisionTreeClassifier(criterion='entropy', max_depth=deep)
    arbol.fit(X_train, y_train)
    train_prec.append(arbol.score(X_train, y_train))
    eval_prec.append(arbol.score(X_test, y_test))

# graficar los resultados.
plt.plot(max_deep_list, train_prec, color='r', label='train')
plt.plot(max_deep_list, eval_prec, color='b', label='test')
plt.title('Grafico de ajuste arbol de decision')
plt.legend()
plt.ylabel('Precision')
plt.xlabel('Cantidad de Nodos')
plt.show()
```

Con este gr치fico de ajuste, se visualiza la precisi칩n del modelo en funci칩n de su complejidad. Aqui podemos ver que el punto con mayor precisi칩n, en los datos de evaluaci칩n, lo obtenemos con un nivel de profundidad de aproximadamente 8 nodos; a partir de all칤 el modelo pierde en generalizaci칩n y comienza a estar sobreajustado
